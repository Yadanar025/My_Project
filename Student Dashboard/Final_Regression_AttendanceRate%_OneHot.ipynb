```python
#import necessary file
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression, LogisticRegression,Ridge, Lasso
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.model_selection import train_test_split,KFold, StratifiedKFold,cross_val_score, KFold,GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score,mean_squared_error
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
attendance_rate = pd.read_csv("cleaned_attendance_rate.csv")
attendance_rate.info()
```

```python
attendance_rate.columns
```

```python

#create X and y

X_att = attendance_rate.drop(columns="Rate_%")
y_att = attendance_rate["Rate_%"]
print(X_att.shape)
print(y_att.shape)
```

```python
X_att_train, X_att_test, y_att_train, y_att_test = train_test_split(X_att, y_att, test_size=0.3, stratify=y_att, random_state=42)
print(X_att_train.info())
print(X_att_test.info())
```

```python
#linear regression
lr_att = LinearRegression()
lr_att.fit(X_att_train,y_att_train)
y_att_pred_lr = lr_att.predict(X_att_test)
print("Linear Regression Model")
print("Predictions: {}, Actual Values: {}".format(y_att_pred_lr[:2], y_att_test[:2]))
print(mean_squared_error(y_att_test,y_att_pred_lr))
print(mean_squared_error(y_att_test,y_att_pred_lr)**(1/2))
#act_vs_test =pd.DataFrame()
#act_vs_test["predict"]=y_att_pred_lr
#act_vs_test["actual"]=y_att_test
#print(act_vs_test)

#decision tree
dt_att = DecisionTreeRegressor(max_depth=6,min_samples_leaf=0.05,random_state=42)
dt_att.fit(X_att_train,y_att_train)
y_att_pred_dt=dt_att.predict(X_att_test)
print("Decision Tree Model")
print("Predictions: {}, Actual Values: {}".format(y_att_pred_dt[:2], y_att_test[:2]))
print(mean_squared_error(y_att_test,y_att_pred_dt))
print(mean_squared_error(y_att_test,y_att_pred_dt)**(1/2))

#random forest
rf_att = RandomForestRegressor()
rf_att.fit(X_att_train,y_att_train)
y_att_pred_rf=rf_att.predict(X_att_test)
print("Random Forest Model")
print("Predictions: {}, Actual Values: {}".format(y_att_pred_rf[:2], y_att_test[:2]))
print(mean_squared_error(y_att_test,y_att_pred_rf))
print(mean_squared_error(y_att_test,y_att_pred_rf)**(1/2))
```

```python
#cross validation
kf = KFold(n_splits = 5, shuffle=True, random_state = 42)
cv_scores_lr = cross_val_score(lr_att,X_att,y_att,cv=kf)
print(f"linear regression : {cv_scores_lr}")

cv_scores_dt = cross_val_score(dt_att,X_att,y_att,cv=kf)
print(f"Decision Tree : {cv_scores_dt}")


cv_scores_rf = cross_val_score(rf_att,X_att,y_att,cv=kf)
print(f"Random forest : {cv_scores_rf}")
```

```python
#cross validation
kf = KFold(n_splits = 5, shuffle=True, random_state = 42)
cv_scores_lr = cross_val_score(lr_att,X_att,y_att,cv=kf,scoring="neg_mean_absolute_error")
print(f"linear regression : {cv_scores_lr}")

cv_scores_dt = cross_val_score(dt_att,X_att,y_att,cv=kf,scoring="neg_mean_absolute_error")
print(f"Decision Tree : {cv_scores_dt}")


cv_scores_rf = cross_val_score(rf_att,X_att,y_att,cv=kf,scoring="neg_mean_absolute_error")
print(f"Random forest : {cv_scores_rf}")
```

```python
#for linear regression
# Define parameter grid
param_grid_att_lr = {'alpha': [0.001, 0.01, 0.1, 1, 10, 100]}

# Ridge Regression
ridge_att_lr = GridSearchCV(Ridge(), param_grid_att_lr, cv=5, scoring="neg_mean_squared_error", n_jobs=-1)
ridge_att_lr.fit(X_att, y_att)
print("Best Ridge Alpha:", ridge_att_lr.best_params_)
print("Best Ridge Score:", ridge_att_lr.best_score_)

# Lasso Regression
lasso_att_lr = GridSearchCV(Lasso(), param_grid_att_lr, cv=5, scoring="neg_mean_squared_error", n_jobs=-1)
lasso_att_lr.fit(X_att, y_att)
print("Best Lasso Alpha:", lasso_att_lr.best_params_)
print("Best Lasso Score:", lasso_att_lr.best_score_)

```

```python
#for best feature with lasso
lasso_att_lr_best = Lasso(alpha=0.001).fit(X_att, y_att)
feature_importance = pd.Series(lasso_att_lr_best.coef_, index=X_att.columns)
print(feature_importance.sort_values())
```

```python
#with standard scaler
scaler = StandardScaler()
X_att_scaled = scaler.fit_transform(X_att)
X_test_att_scaled = scaler.transform(X_att_test)
lasso_att_lr_best = Lasso(alpha=0.001).fit(X_att_scaled, y_att)
feature_importance = pd.Series(lasso_att_lr_best.coef_, index=X_att.columns)
print(feature_importance.sort_values())
```

```python
y_att_pred_lasso = lasso_att_lr_best.predict(X_test_att_scaled)
print("Lasso")
print("Predictions: {}, Actual Values: {}".format(y_att_pred_lasso[:2], y_att_test[:2]))
print(mean_squared_error(y_att_test,y_att_pred_lasso))
print(mean_squared_error(y_att_test,y_att_pred_lasso)**(1/2))
```

```python
#Hyperparameter Tuning for Random Forest

param_grid_rf_att =  {
    "n_estimators": [5, 50, 100, 200],
    "max_depth": [1,5,10,None],
    "min_samples_split": [2, 5, 10, None],
    "min_samples_leaf": [ 2, 4,10]
}
grid_search_rf_att = GridSearchCV(RandomForestRegressor(random_state=42), param_grid_rf_att, cv=5, scoring="neg_mean_squared_error")
grid_search_rf_att.fit(X_att, y_att)

print("Best Parameters:", grid_search_rf_att.best_params_)
print("Best Score:", grid_search_rf_att.best_score_)
```
